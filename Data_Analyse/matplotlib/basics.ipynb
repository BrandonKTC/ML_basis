```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
plt.plot(x,y);
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xlim(0,6)
plt.ylim(0,18)
plt.title('String Title')
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
x = np.arange(0,10)
```

```python
y = x*2
```

```python
fig = plt.figure()

# LARGE AXES

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# SMALL AXES

axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(x,y)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

```python
fig = plt.figure(figsize=(2,2),dpi=300)

axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('new_figure.png', bbox_inches='tight')
```

```python
fig, axes = plt.subplots(ncols=2)

axes[0].plot(x,y)
axes[1].plot(a,b)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[0][1].plot(a,b)
axes[1][0].plot(a,b)
axes[1][1].plot(x,y)

fig.subplots_adjust(wspace=0.9, hspace=0.5)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x, color="purple", label='X vs X')
ax.plot(x,x**2, color="#000", label='X vs X**2')

ax.legend()
```
