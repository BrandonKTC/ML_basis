# MERGE

### INNER MERGE

```python
registration = pd.DataFrame({'reg_id': [1,2,3,4], 'name': ['Andrew', 'Bobo', 'Claire', 'David']})
logins = pd.DataFrame({'log_id': [1,2,3,4], 'name':['Xavier', 'Andrew', 'Yolanda', 'Bobo']})
```

```python
registration
```

```python
logins
```

```python
pd.merge(registration, logins, how='inner', on='name')
```

### LEFT & RIGHT MERGE

```python
pd.merge(registration, logins, how='left', on='name')
```

```python
pd.merge(registration, logins, how="right", on='name')
```

### OUTER MERGE

```python
pd.merge(registration, logins, how="outer", on="name")
```

```python
registration.set_index("name", inplace=True)
```

```python
registration
```

```python
logins
```

```python
pd.merge(registration, logins, left_index=True, right_on="name")
```

```python
registration.reset_index(inplace=True)
```

```python
registration.columns = ['name','id']
```

```python
logins.columns = ['id','name']
```

```python
pd.merge(registration, logins, how='inner', on="name")
```

```python
pd.merge(registration, logins, how='inner', on="name", suffixes=('_reg','_log'))
```
