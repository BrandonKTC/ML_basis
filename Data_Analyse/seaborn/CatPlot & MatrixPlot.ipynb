```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.catplot(data=df, x='gender', y='math score',kind='box')
```

```python
sns.catplot(data=df, x='gender', y='math score',kind='box', row="test preparation course")
```

```python
sns.catplot(data=df, x='gender', y='math score',kind='box', row="test preparation course", col='lunch')
```

```python
sns.catplot(data=df, x='gender', y='math score',kind='violin', row="test preparation course", col='lunch')
```

## MatrixPlot

```python
df = pd.read_csv('country_table.csv')
```

```python
df.head()
```

```python
df = df.set_index("Countries")
```

```python
# we should have a label index to be able to use the matrix plot
df
```

```python
sns.heatmap(df.drop('Life expectancy', axis=1),lw=.5)
```

```python
sns.heatmap(df.drop('Life expectancy', axis=1),lw=.5, annot=True)
```

```python
sns.heatmap(df.drop('Life expectancy', axis=1),lw=.5,annot=True, cmap="viridis")
```

```python
sns.clustermap(df.drop('Life expectancy', axis=1),lw=.5,annot=True, cmap="viridis")
```

```python
sns.clustermap(df.drop('Life expectancy', axis=1),lw=.5,annot=True, cmap="viridis", col_cluster=False)
```
