## Part One: SVM - Classification

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../DATA/mouse_viral_study.csv')
```

```python
df.head()
```

```python
sns.scatterplot(x="Med_1_mL", y="Med_2_mL", hue='Virus Present', data=df)

# HYPERPLANE (2d line)

x = np.linspace(0,10,100)
m = -1
b = 11
y = m*x +b
plt.plot(x,y,'black');
```

```python
from sklearn.svm import SVC
```

```python
# help(SVC)
```

```python
X = df.drop('Virus Present',axis=1)
y = df['Virus Present']
```

```python
model = SVC(kernel='linear', C=1000)
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

```python
plot_svm_boundary(model, X, y)
```

# Part Two: Hyperparameters

```python
model = SVC(kernel='linear', C=.05)
model.fit(X,y)
```

**C parameter**

```python
# as C get close to 0 we allow more point into the margin
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='linear', C=.005)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=.1)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=.001)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

**gamma parameter**

```python
model = SVC(kernel='rbf', C=1, gamma="scale")
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma="auto")
model.fit(X,y)
```

```python
# pick up a lot more noise (overfit to the data)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='rbf', C=1, gamma=2)
model.fit(X,y)
```

```python
plot_svm_boundary(model,X,y)
```

**kernel**

```python
model = SVC(kernel='sigmoid')
model.fit(X,y)
```

```python
# (for this dataset) it rely on point at the center to have information of where the margin should go
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='poly', C=1, degree=1)
model.fit(X,y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='poly', C=1, degree=2)
model.fit(X,y)
plot_svm_boundary(model, X, y)
```

```python
model = SVC(kernel='poly', C=.005, degree=5)
model.fit(X,y)
plot_svm_boundary(model, X, y)
```

```python
# Grid Search
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid = {"C":[.01,.1,1], 
              'kernel':['poly','sigmoid','rbf','linear']}
```

```python
grid = GridSearchCV(svm, param_grid)
```

```python
grid.fit(X,y)
```
