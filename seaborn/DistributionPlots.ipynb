```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

## RugPlot

```python
sns.rugplot(x="salary", data=df)
```

```python
sns.rugplot(x="salary", data=df, height=0.5)
```

```python
plt.figure(figsize=(6,2), dpi=300)
sns.rugplot(x="salary", data=df, height=0.5)
```

## HistogramPlot

```python
sns.displot(data=df, x="salary", bins=20, rug=True)
```

```python
sns.displot(data=df, x="salary")
```

```python
sns.displot(data=df, x="salary", bins=100)
```

```python
sns.set(style="whitegrid")
sns.displot(data=df, x="salary", bins=20)
```

```python
sns.set(style="darkgrid")
sns.displot(data=df, x="salary", bins=20)
```

```python
sns.histplot(data=df, x="salary")
```

```python
sns.displot(data=df, x="salary", bins=20, kde=True)
```

```python
plt.figure(dpi=300)
sns.set(style="white")
sns.displot(data=df, x="salary", kde=True, rug=True)
```

## KDEPlot

```python
sns.kdeplot(data=df, x="salary")
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=["age"])
```

```python
sns.rugplot(sample_ages)
```

```python
sns.displot(sample_ages, x="age", rug=True, bins=30)
```

```python
sns.displot(sample_ages, x="age", rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x="age", clip=[0,100])
```

```python
sns.kdeplot(data=sample_ages, x="age", clip=[0,100], bw_adjust=.1)
```

```python
sns.kdeplot(data=sample_ages, x="age", clip=[0,100], bw_adjust=.6, fill=True)
```
