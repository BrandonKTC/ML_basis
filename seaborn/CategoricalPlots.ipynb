```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

## CountPlot

```python
plt.figure(figsize=(10,4), dpi=200)
sns.countplot(data=df, x="division")
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, x="level of education")
```

```python
plt.figure(figsize=(10,4))
sns.countplot(data=df, x="level of education", hue="division")
```

## BarPlot

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, y="salary", x="level of education")
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, y="salary", x="level of education", estimator=np.mean)
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, y="salary", x="level of education", errorbar='sd')
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=df, y="salary", x="level of education", errorbar='sd', hue="division")

plt.legend(bbox_to_anchor=(1.05,1))
```

# PART 2

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

## BoxPlot

```python
sns.boxplot(data=df, y="math score")
```

```python
sns.boxplot(data=df, y="math score", x="test preparation course")
```

```python
sns.boxplot(data=df, y="reading score", x="test preparation course")
```

```python
plt.figure(figsize=(10,4), dpi=300)
sns.boxplot(data=df, y="reading score",x="parental level of education", hue="test preparation course")
```

```python
plt.figure(figsize=(10,4), dpi=300)
sns.boxplot(data=df, x="reading score",y="parental level of education", hue="test preparation course", palette="Set2")

plt.legend(bbox_to_anchor=(1.05,.5))
```

## ViolinPlot

```python
plt.figure(figsize=(10,6), dpi=300)
sns.violinplot(data=df, x="reading score",y="parental level of education", hue="test preparation course", palette="Set2")

plt.legend(bbox_to_anchor=(1.55,.5))
```

```python
plt.figure(figsize=(10,6), dpi=300)
sns.violinplot(data=df, x="reading score",y="parental level of education", hue="test preparation course", palette="Set2", split=True)

plt.legend(bbox_to_anchor=(1.55,.5))
```

```python
plt.figure(figsize=(10,6), dpi=300)
sns.violinplot(data=df, x="reading score",y="parental level of education", hue="test preparation course", palette="Set2", split=True, inner=None)

plt.legend(bbox_to_anchor=(1.55,.5))
```

```python
plt.figure(figsize=(10,6), dpi=300)
sns.violinplot(data=df, x="reading score",y="parental level of education", hue="test preparation course", palette="Set2", split=True, inner="quartile")

plt.legend(bbox_to_anchor=(1.55,.5))
```

```python
plt.figure(figsize=(10,6), dpi=300)
sns.violinplot(data=df, x="reading score",y="parental level of education", split=True,)
```

## SwarmPlot

```python
sns.swarmplot(data=df, x='math score', size=2)
```

```python
sns.swarmplot(data=df, x='math score', y="gender", size=2)
```

```python
sns.swarmplot(data=df, x='math score', y="gender", hue="test preparation course", size=2)
```

```python
sns.swarmplot(data=df, x='math score', y="gender", hue="test preparation course", size=2,dodge=True)
```

## BoxenPlot

```python
sns.boxenplot(data=df, x='math score', hue="gender", y="test preparation course")
```
